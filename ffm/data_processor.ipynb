# 第一步：获取部分字段的取值有几种

```python
# for site_id, site_domain, app_id, app_domain, device_model,
# device_ip, device_id fields,C14,C17,C19,C21, one-hot using frequency
# for other fields, one-hot-encoding directly

train = pd.read_csv('./train.csv', chunksize=10000)

# one-hot encoding directly
click = set()
hour = set()
C1 = set()
banner_pos = set()
site_category = set()
app_category = set()
device_type = set()
device_conn_type = set()
C15 = set()
C16 = set()
C18 = set()
C20 = set()

hour = set(range(24))

# one-encoding by frequency bucket
C14 = []
C17 = []
C19 = []
C21 = []
site_id = []
site_domain = []
app_id = []
app_domain = []
device_model = []
device_ip = []
device_id = []




for data in train:

    click_v = set(data['click'].values)
    click = click | click_v

    C1_v = set(data['C1'].values)
    C1 = C1 | C1_v

    C15_v = set(data['C15'].values)
    C15 = C15 | C15_v

    C16_v = set(data['C16'].values)
    C16 = C16 | C16_v

    C18_v = set(data['C18'].values)
    C18 = C18 | C18_v

    C20_v = set(data['C20'].values)
    C20 = C20 | C20_v

    banner_pos_v = set(data['banner_pos'].values)
    banner_pos = banner_pos | banner_pos_v

    site_category_v = set(data['site_category'].values)
    site_category = site_category | site_category_v

    app_category_v = set(data['app_category'].values)
    app_category = app_category | app_category_v

    device_type_v = set(data['device_type'].values)
    device_type = device_type | device_type_v

    device_conn_type_v = set(data['device_conn_type'].values)
    device_conn_type = device_conn_type | device_conn_type_v

print('click: ', click)
print('C1: ', C1)
print('C15: ', C15)
print('C16: ', C16)
print('C18: ', C18)
print('C20: ', C20)
print('banner_pos: ', banner_pos)
print('site_category: ', site_category)
print('app_category: ', app_category)
print('device_type: ', device_type)
print('device_conn_type: ', device_conn_type)
# save dictionaries
with open('./data/sets/click.pkl','wb') as f:
    pickle.dump(click,f)

with open('./data/sets/hour.pkl','wb') as f:
    pickle.dump(hour,f)

with open('./data/sets/C1.pkl','wb') as f:
    pickle.dump(C1,f)

with open('./data/sets/C15.pkl','wb') as f:
    pickle.dump(C15,f)

with open('./data/sets/C16.pkl','wb') as f:
    pickle.dump(C16,f)

with open('./data/sets/C18.pkl','wb') as f:
    pickle.dump(C18,f)

with open('./data/sets/C20.pkl','wb') as f:
    pickle.dump(C20,f)

with open('./data/sets/banner_pos.pkl','wb') as f:
    pickle.dump(banner_pos,f)

with open('./data/sets/site_category.pkl','wb') as f:
    pickle.dump(site_category,f)

with open('./data/sets/app_category.pkl','wb') as f:
    pickle.dump(app_category,f)

with open('./data/sets/device_type.pkl','wb') as f:
    pickle.dump(device_type,f)

with open('./data/sets/device_conn_type.pkl','wb') as f:
    pickle.dump(device_conn_type,f)
```

# 计算每个取值的个数

```python
train = pd.read_csv('./train.csv', chunksize=10000)

C14 = dict()
C17 = dict()
C19 = dict()
C21 = dict()
site_id = dict()
site_domain = dict()
app_id = dict()
app_domain = dict()
device_model = dict()
device_id = dict()
device_ip = dict()

count = 0
for data in train:
    C14_list = data['C14'].values
    for k,v in Counter(C14_list).items():
        if k in C14.keys():
            C14[k] += v
        else:
            C14[k] = v

    C17_list = data['C17'].values
    for k,v in Counter(C17_list).items():
        if k in C17.keys():
            C17[k] += v
        else:
            C17[k] = v

    C19_list = data['C19'].values
    for k,v in Counter(C19_list).items():
        if k in C19.keys():
            C19[k] += v
        else:
            C19[k] = v

    C21_list = data['C21'].values
    for k,v in Counter(C21_list).items():
        if k in C21.keys():
            C21[k] += v
        else:
            C21[k] = v

    site_id_list = data['site_id'].values
    for k,v in Counter(site_id_list).items():
        if k in site_id.keys():
            site_id[k] += v
        else:
            site_id[k] = v

    site_domain_list = data['site_domain'].values
    for k,v in Counter(site_domain_list).items():
        if k in site_domain.keys():
            site_domain[k] += v
        else:
            site_domain[k] = v

    app_id_list = data['app_id'].values
    for k,v in Counter(app_id_list).items():
        if k in app_id.keys():
            app_id[k] += v
        else:
            app_id[k] = v

    app_domain_list = data['app_domain'].values
    for k,v in Counter(app_domain_list).items():
        if k in app_domain.keys():
            app_domain[k] += v
        else:
            app_domain[k] = v

    device_model_list = data['device_model'].values
    for k,v in Counter(device_model_list).items():
        if k in device_model.keys():
            device_model[k] += v
        else:
            device_model[k] = v

    device_id_list = data['device_id'].values
    for k,v in Counter(device_id_list).items():
        if k in device_id.keys():
            device_id[k] += v
        else:
            device_id[k] = v

    device_ip_list = data['device_ip'].values
    for k,v in Counter(device_ip_list).items():
        if k in device_ip.keys():
            device_ip[k] += v
        else:
            device_ip[k] = v

    count += 1
    if count % 100 == 0:
        print('{} has finished'.format(count))
            
print('C14: ', C14 )
print('C17: ', C17 )
print('C19: ', C19 )
print('C21: ', C21 )
print('site_id: ', site_id )
print('site_domain: ', site_domain )
print('app_id: ', app_id )
print('app_domain: ', app_domain )
print('device_model: ', device_model )
print('device_id: ', device_id )
print('device_ip: ', device_ip )

# save dictionaries
with open('./data/field2count/C14.pkl','wb') as f:
    pickle.dump(C14,f)

with open('./data/field2count/C17.pkl','wb') as f:
    pickle.dump(C17,f)

with open('./data/field2count/C19.pkl','wb') as f:
    pickle.dump(C19,f)

with open('./data/field2count/C21.pkl','wb') as f:
    pickle.dump(C21,f)

with open('./data/field2count/site_id.pkl','wb') as f:
    pickle.dump(site_id,f)

with open('./data/field2count/site_domain.pkl','wb') as f:
    pickle.dump(site_domain,f)

with open('./data/field2count/app_id.pkl','wb') as f:
    pickle.dump(app_id,f)

with open('./data/field2count/app_domain.pkl','wb') as f:
    pickle.dump(app_domain,f)

with open('./data/field2count/device_model.pkl','wb') as f:
    pickle.dump(device_model,f)

with open('./data/field2count/device_id.pkl','wb') as f:
    pickle.dump(device_id,f)

with open('./data/field2count/device_ip.pkl','wb') as f:
    pickle.dump(device_ip,f)
```

```python
direct_encoding_fields = ['hour', 'C1', 'C15', 'C16', 'C18', 'C20',
                          'banner_pos',  'site_category','app_category',
                          'device_type','device_conn_type']

frequency_encoding_fields = ['C14','C17', 'C19', 'C21',
                             'site_id','site_domain','app_id','app_domain',
                              'device_model', 'device_id']

# load direct encoding fields
with open('./data/sets/click.pkl','rb') as f:
    click = pickle.load(f)

with open('./data/sets/hour.pkl','rb') as f:
    hour = pickle.load(f)

with open('./data/sets/C1.pkl','rb') as f:
    C1 = pickle.load(f)

with open('./data/sets/C15.pkl','rb') as f:
    C15 = pickle.load(f)

with open('./data/sets/C16.pkl','rb') as f:
    C16 = pickle.load(f)

with open('./data/sets/C18.pkl','rb') as f:
    C18 = pickle.load(f)

with open('./data/sets/C20.pkl','rb') as f:
    C20 = pickle.load(f)

with open('./data/sets/banner_pos.pkl','rb') as f:
    banner_pos = pickle.load(f)

with open('./data/sets/site_category.pkl','rb') as f:
    site_category = pickle.load(f)

with open('./data/sets/app_category.pkl','rb') as f:
    app_category = pickle.load(f)

with open('./data/sets/device_type.pkl','rb') as f:
    device_type = pickle.load(f)

with open('./data/sets/device_conn_type.pkl','rb') as f:
    device_conn_type = pickle.load(f)


# loading frequency encoding fields
# field2count dictionaries
with open('./data/field2count/C14.pkl','rb') as f:
    C14 = pickle.load(f)

with open('./data/field2count/C17.pkl','rb') as f:
    C17 = pickle.load(f)

with open('./data/field2count/C19.pkl','rb') as f:
    C19 = pickle.load(f)

with open('./data/field2count/C21.pkl','rb') as f:
    C21 = pickle.load(f)

with open('./data/field2count/site_id.pkl','rb') as f:
    site_id = pickle.load(f)

with open('./data/field2count/site_domain.pkl','rb') as f:
    site_domain = pickle.load(f)

with open('./data/field2count/app_id.pkl','rb') as f:
    app_id = pickle.load(f)

with open('./data/field2count/app_domain.pkl','rb') as f:
    app_domain = pickle.load(f)

with open('./data/field2count/device_model.pkl','rb') as f:
    device_model = pickle.load(f)

with open('./data/field2count/device_id.pkl','rb') as f:
    device_id = pickle.load(f)
# 每个特征取值对应的下标
field_dict = {}
# feature和field的对应关系
feature2field = {}
# field的下标
field_index = 0
ind = 0
# 对于离散特征，直接one-hot
for field in direct_encoding_fields:
    # value to one-hot-encoding index dict
    field_sets = eval(field)
    for value in list(field_sets):
        field_dict[value] = ind
        feature2field[ind] = field_index
        ind += 1
    field_index += 1
    with open('./data/dicts/'+field+'.pkl', 'wb') as f:
        pickle.dump(field_dict, f)
# 对于连续特征，
for field in frequency_encoding_fields:
    # value to one-hot-encoding index dict
    field2count = eval(field)
    index_rare = None
    for k,count in field2count.items():
        if count < 10:
            if index_rare == None:
                field_dict[k] = ind
                feature2field[ind] = field_index
                index_rare = ind
                ind += 1
            else:
                field_dict[k] = index_rare
                feature2field[index_rare] = field_index

        else:
            field_dict[k] = ind
            feature2field[ind] = field_index
            ind += 1
    field_index += 1
    with open('./data/dicts/'+field+'.pkl', 'wb') as f:
        pickle.dump(field_dict, f)



field_dict = {}
field_sets = click
for value in list(field_sets):
    field_dict[value] = ind
    ind += 1
print('field_dict ', field_dict)
print('feature2field', feature2field)

with open('./data/dicts/'+'click'+'.pkl', 'wb') as f:
    pickle.dump(field_dict, f)



with open('./data/feature2field.pkl', 'wb') as f:
    pickle.dump(feature2field, f)
```
